```python
# =============================
# Part 1: Loan Status Classification - Running Code
# =============================

# Step 0: Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, confusion_matrix
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
import os

# Step 1: Load Training Data
df = pd.read_csv("train.csv")

# Step 2: Quick Look at Target
sns.countplot(data=df, x="loan_status")
plt.show()

# Step 3: Prepare Features and Target
X = df.drop('loan_status', axis=1)  # Features
y = df['loan_status']               # Target

# Step 4: Identify numeric and categorical columns
numeric_features = X.select_dtypes(include=['int64','float64']).columns.tolist()
categorical_features = X.select_dtypes(include=['object']).columns.tolist()

# Step 5: Build Preprocessing Pipelines
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(drop='first', handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)

# Step 6: Combine Preprocessing with Logistic Regression
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(max_iter=1000))
])

# Step 7: Train the model
model.fit(X, y)

# Step 8: Load Test Data (Create small test if missing)
if not os.path.exists("test.csv"):
    print("test.csv not found. Creating a small test set from training data...")
    df.sample(20).to_csv("test.csv", index=False)

d_test = pd.read_csv("test.csv")
xtest = d_test.drop('loan_status', axis=1)
ytest = d_test['loan_status']

# Step 9: Ensure labels are consistent for F1 score
# Convert to strings only if the target is numeric
if ytest.dtype in ['int64', 'float64']:
    ytest_labels = ytest
    prediction_labels = model.predict(xtest)
else:
    ytest_labels = ytest.astype(str)
    prediction_labels = pd.Series(model.predict(xtest)).astype(str)

# Step 10: Evaluate Model
print("F1 Score:", f1_score(ytest_labels, prediction_labels))
conf_matrix = confusion_matrix(ytest_labels, prediction_labels)
sns.heatmap(conf_matrix, annot=True, fmt='d')
plt.show()

# =============================
# Step 11: Notes 
# =============================
# - Numeric missing values filled with median
# - Categorical missing values filled with most frequent
# - Categorical features one-hot encoded
# - Numeric features scaled
# - Logistic Regression classifier
# - Pipeline allows predictions on new datasets with same columns
```
